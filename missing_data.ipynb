Her bir değişken için eksik değerleri nasıl doldurabileceğinizi düşünün. Eksik değerleri bir değerle doldurmak hangi değişkenler için anlamlı, hangileri için anlamsızdır?

```python
for state in states.STATE:
    local_revenue=states.loc[states['STATE'] == state, 'LOCAL_REVENUE'].mean()
    #print(local_revenue)
    states["LOCAL_REVENUE"].fillna(local_revenue, inplace=True)
```

```python
for state in states.STATE:
    states["STATE_REVENUE"].fillna(states.loc[states['STATE'] == state, 'STATE_REVENUE'].mean(), inplace=True)
```

```python
for state in states.STATE:
    states["FEDERAL_REVENUE"].fillna(states.loc[states['STATE'] == state, 'FEDERAL_REVENUE'].mean(), inplace=True)
```

```python
for state in states.STATE:
    #total=states.loc[states['STATE'] == state, ['FEDERAL_REVENUE','STATE_REVENUE']].sum(axis=1)
    states["TOTAL_REVENUE"].fillna(states.loc[states['STATE'] == state, ['FEDERAL_REVENUE','STATE_REVENUE']].sum(axis=1), inplace=True)
```

```python
for state in states.STATE:
    states["CAPITAL_OUTLAY_EXPENDITURE"].fillna(states.loc[states['STATE'] == state, 'CAPITAL_OUTLAY_EXPENDITURE'].mean(), inplace=True)
    states["OTHER_EXPENDITURE"].fillna(states.loc[states['STATE'] == state, 'OTHER_EXPENDITURE'].mean(), inplace=True)
    states["SUPPORT_SERVICES_EXPENDITURE"].fillna(states.loc[states['STATE'] == state, 'SUPPORT_SERVICES_EXPENDITURE'].mean(), inplace=True)
    states["INSTRUCTION_EXPENDITURE"].fillna(states.loc[states['STATE'] == state, 'INSTRUCTION_EXPENDITURE'].mean(), inplace=True)
    #states["TOTAL_EXPENDITURE"].fillna(states.loc[states['STATE'] == state, 'CAPITAL_OUTLAY_EXPENDITURE','OTHER_EXPENDITURE','SUPPORT_SERVICES_EXPENDITURE','INSTRUCTION_EXPENDITURE'].sum(axis=1), inplace=True)
```

```python
for state in states.STATE:
    total_first=states.loc[states['STATE'] == state, ['CAPITAL_OUTLAY_EXPENDITURE','OTHER_EXPENDITURE']].sum(axis=1)
    total_second=states.loc[states['STATE'] == state, ['SUPPORT_SERVICES_EXPENDITURE','INSTRUCTION_EXPENDITURE']].sum(axis=1)
    total=total_first+total_second
    states["TOTAL_EXPENDITURE"].fillna(total, inplace=True)
```

## Şimdi zaman faktörünü dikkate alma zamanı! 2. sorudaki cevabınızı tekrar gözden geçirin ve eksik verileri o yıl içerisinde gözlemlenen değerlere dayanarak doldurun. Örneğin, bir değeri ortalama değer ile doldurmak isterseniz, o yılın ortalamasını hesaplayın.

```python
import pandas as pd
import numpy as np
states=pd.read_csv('data/states_all.csv')
states.isnull().sum()*100/states.count()
```

```python
for year in states.YEAR.unique():
    for state in states.STATE:
        local_revenue=states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'LOCAL_REVENUE'].mean()
        states["LOCAL_REVENUE"].fillna(local_revenue, inplace=True)
```

```python
states.isnull().sum()*100/states.count()
```

```python
for year in states.YEAR.unique():   
    for state in states.STATE:
        states["STATE_REVENUE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'STATE_REVENUE'].mean(), inplace=True)
        states["FEDERAL_REVENUE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'FEDERAL_REVENUE'].mean(), inplace=True)
```

```python
for year in states.YEAR.unique():  
    for state in states.STATE:
        states["TOTAL_REVENUE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), ['FEDERAL_REVENUE','STATE_REVENUE']].sum(axis=1), inplace=True)
```

```python
for year in states.YEAR.unique(): 
    for state in states.STATE:
        states["CAPITAL_OUTLAY_EXPENDITURE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'CAPITAL_OUTLAY_EXPENDITURE'].mean(), inplace=True)
        states["OTHER_EXPENDITURE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'OTHER_EXPENDITURE'].mean(), inplace=True)
        states["SUPPORT_SERVICES_EXPENDITURE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'SUPPORT_SERVICES_EXPENDITURE'].mean(), inplace=True)
        states["INSTRUCTION_EXPENDITURE"].fillna(states.loc[(states['YEAR'] == year) & (states['STATE'] == state), 'INSTRUCTION_EXPENDITURE'].mean(), inplace=True)
```

```python
for year in states.YEAR.unique():    
    for state in states.STATE:
        total_first=states.loc[(states['YEAR'] == year) & (states['STATE'] == state), ['CAPITAL_OUTLAY_EXPENDITURE','OTHER_EXPENDITURE']].sum(axis=1)
        total_second=states.loc[(states['YEAR'] == year) & (states['STATE'] == state), ['SUPPORT_SERVICES_EXPENDITURE','INSTRUCTION_EXPENDITURE']].sum(axis=1)
        total=total_first+total_second
        states["TOTAL_EXPENDITURE"].fillna(total, inplace=True)
```

```python
for year in states.YEAR.unique(): 
    states["GRADES_1_8_G"].fillna(states.loc[states['YEAR'] == year, "GRADES_1_8_G"].mean(), inplace=True)
    states["GRADES_9_12_G"].fillna(states.loc[states['YEAR'] == year, "GRADES_9_12_G"].mean(), inplace=True)
    states["AVG_MATH_4_SCORE"].fillna(states.loc[states['YEAR'] == year, "AVG_MATH_4_SCORE"].mean(), inplace=True)
    states["AVG_MATH_8_SCORE"].fillna(states.loc[states['YEAR'] == year, "AVG_MATH_8_SCORE"].mean(), inplace=True)
    states["AVG_READING_4_SCORE"].fillna(states.loc[states['YEAR'] == year, "AVG_READING_4_SCORE"].mean(), inplace=True)
    states["AVG_READING_8_SCORE"].fillna(states.loc[states['YEAR'] == year, "AVG_READING_8_SCORE"].mean(), inplace=True)
```

```python
states.isnull().sum()*100/states.count()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(40,20))
y=len(states.columns)
for i in range(3,y):
    plt.subplot(5,5,i-2)
    plt.bar(states.index,states[states.columns[i]],color = "red")
    plt.title(states.columns[i])
plt.tight_layout()
plt.show()
```

## Bu sefer, eksik değerleri enterpolasyon yaparak doldurun.

```python
import pandas as pd
import numpy as np
states=pd.read_csv('data/states_all.csv')
states.isnull().sum()*100/states.count()
```

```python
states.head()
```

```python
states_sorted=states.sort_values('YEAR')
new_states=states_sorted.interpolate(interplace=True)
new_states.isnull().sum()*100/new_states.count()
#new_states.head()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(40,20))
y=len(states.columns)
for i in range(3,y):
    plt.subplot(5,5,i-2)
    plt.bar(states.index,states[states.columns[i]],color = "black")
    plt.title(states.columns[i])
plt.tight_layout()
plt.show()
```
